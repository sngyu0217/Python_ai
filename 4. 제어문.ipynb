## while 

````python
day=1

while day<=5:
  hour=1

  while hour<=3:
    print(f"{day} {hour}.")
    hour+=1
  day+=1
````

````python
print(day,hour)
````

````python
# 5      60     
#          

# 1) 
# 2)  
# 3)  

scores=[90,25,67,45,80]
number=0

for score in scores:
    number+=1

    if score>60:
        print(f"{number}  ")

    else:
        print(f"{number}  ")
````

````python
idx=0
while idx<len(scores):
    if scores[idx]>=60:
        print("")
    else:
        print("")
    idx+=1
````
